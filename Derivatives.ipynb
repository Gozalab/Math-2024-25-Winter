## 21. Derivatives

### 1. Compute derivatives of functions:

- $y(x) = -3x + 3$
$$ y'(x) = \frac{d}{dx} (-3x + 3) = -3 $$

- $y(x) = \pi x + \sin(1)$
$$ y'(x) = \frac{d}{dx} (\pi x + \sin(1)) = \pi $$

- $y(x) = 4 + \sin(2)$
$$ y'(x) = \frac{d}{dx} (4 + \sin(2)) = 0 $$

- $y(x) = 2x^3 - 3x^2 + 8x - 9$
$$ y'(x) = \frac{d}{dx} (2x^3 - 3x^2 + 8x - 9) = 6x^2 - 6x + 8 $$

- $y(x) = 6x^{1/3}$
$$ y'(x) = \frac{d}{dx} (6x^{1/3}) = 2x^{-2/3} $$

- $y(x) = \sqrt{x}$
$$ y'(x) = \frac{d}{dx} (\sqrt{x}) = \frac{1}{2\sqrt{x}} $$

- $y(x) = \cos(x) + \sin(x)$
$$ y'(x) = \frac{d}{dx} (\cos(x) + \sin(x)) = -\sin(x) + \cos(x) $$

- $y(x) = 2\sin(x) \cos(x)$
$$ y'(x) = \frac{d}{dx} (2\sin(x) \cos(x)) = 2(\cos^2(x) - \sin^2(x)) $$

- $y(x) = x \sin(x)$
$$ y'(x) = \frac{d}{dx} (x \sin(x)) = \sin(x) + x \cos(x) $$

- $y(x) = (x+1)(x+1)$
$$ y'(x) = \frac{d}{dx} ((x+1)^2) = 2(x+1) $$

- $y(x) = \frac{x}{x+1}$
$$ y'(x) = \frac{d}{dx} \left( \frac{x}{x+1} \right) = \frac{1}{(x+1)^2} $$

- $y(x) = (x+1)\exp(x)$
$$ y'(x) = \frac{d}{dx} \left( (x+1)\exp(x) \right) = \exp(x)(x+1) + \exp(x) $$

- $y(x) = \sin(x^2)$
$$ y'(x) = \frac{d}{dx} (\sin(x^2)) = 2x \cos(x^2) $$

- $y(x) = \exp(-2x)$
$$ y'(x) = \frac{d}{dx} (\exp(-2x)) = -2\exp(-2x) $$

- $y(x) = \frac{1}{\sin(x+1)}$
$$ y'(x) = \frac{d}{dx} \left( \frac{1}{\sin(x+1)} \right) = -\frac{\cos(x+1)}{\sin^2(x+1)} $$

- $y(x) = \sqrt{2x+1}$
$$ y'(x) = \frac{d}{dx} (\sqrt{2x+1}) = \frac{1}{\sqrt{2x+1}} $$

### 2. Prove:
$$ \frac{d}{dx} (\ln(\sin(x))) = \cot(x) $$

**Solution:**
Using the chain rule:
$$
\frac{d}{dx} (\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x)
$$

**Result:** $ \frac{d}{dx} (\ln(\sin(x))) = \cot(x) $

### 3. For $ f(x) = \cos(x) $, verify that $ f''(x) = -f(x) $.

**Solution:**

First derivative:
$$
f'(x) = -\sin(x)
$$

Second derivative:
$$
f''(x) = -\cos(x)
$$

Since $ f(x) = \cos(x) $, we have:
$$
f''(x) = -f(x)
$$

**Result:** Verified $ f''(x) = -f(x) $

### 4. Using de l'Hospital's Rule, find the improper limits:

- $ \lim_{x \to 0} \frac{\sin{x}}{x} $

**Solution:**
Since this is of the form $ \frac{0}{0} $, apply L'Hopital's Rule:
$$
\lim_{x \to 0} \frac{\sin{x}}{x} = \lim_{x \to 0} \frac{\cos{x}}{1} = 1
$$

**Result:** $ 1 $

---

- $ \lim_{x \to \infty} \frac{\ln(x)}{x} $

**Solution:**
This is of the form $ \frac{\infty}{\infty} $, so apply L'Hopital's Rule:
$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{1/x}{1} = 0
$$

**Result:** $ 0 $

---

- $ \lim_{x \to \infty} \frac{\exp(x)}{x} $

**Solution:**
This is of the form $ \frac{\infty}{\infty} $, so apply L'Hopital's Rule:
$$
\lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\exp(x)}{1} = \infty
$$

**Result:** $ \infty $

### 5. In physics, the position of a particle is given by $ x(t) = 3t^2 - 6t + 1 $. Find the velocity $ V(t) = x'(t) $ and acceleration $ a(t) = V'(t) = x''(t) $ of the particle at time $ t = 2 $.

**Solution:**

- Velocity $ V(t) = x'(t) $:
$$
V(t) = \frac{d}{dt} (3t^2 - 6t + 1) = 6t - 6
$$
At $ t = 2 $:
$$
V(2) = 6(2) - 6 = 12 - 6 = 6
$$

- Acceleration $ a(t) = x''(t) $:
$$
a(t) = \frac{d}{dt} (6t - 6) = 6
$$

**Result:**
- Velocity at $ t = 2 $: $ V(2) = 6 $
- Acceleration: $ a(t) = 6 $
