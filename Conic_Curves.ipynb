## Tasks Related to Conic Sections

### General Notes on Conic Sections:

1. **Circle:**
   - General equation:
     $$
     (x - h)^2 + (y - k)^2 = r^2
     $$
     - \((h, k)\) is the center.
     - \(r\) is the radius.

2. **Parabola:**
   - General equation (vertical axis):
     $$
     y = ax^2 + bx + c
     $$
   - General equation (horizontal axis):
     $$
     x = ay^2 + by + c
     $$
   - The vertex, focus, and directrix determine the shape and orientation.

3. **Ellipse:**
   - Standard equation:
     $$
     \frac{(x-h)^2}{a^2} + \frac{(y-k)^2}{b^2} = 1
     $$
     - \((h, k)\) is the center.
     - \(a\) and \(b\) are the semi-major and semi-minor axes.

4. **Hyperbola:**
   - Standard equation:
     $$
     \frac{(x-h)^2}{a^2} - \frac{(y-k)^2}{b^2} = 1
     $$
     - \((h, k)\) is the center.
     - Asymptotes pass through the center.

### Tasks and Their Solutions:

#### **1. Equation of a circle with center at \(A(1, 2)\) and radius \(r = 3\):**
- Using the circle equation:
  $$
  (x - 1)^2 + (y - 2)^2 = 3^2
  $$
- Final equation:
  $$
  (x - 1)^2 + (y - 2)^2 = 9
  $$

---

#### **2. Equation of a parabola intersecting the \(Ox\) axis at \(x = 2\), \(x = 4\), and passing through \(y(3) = 1\):**
- Parabola equation:
  $$
  y = a(x - 2)(x - 4)
  $$
- Substitute point \((3, 1)\) to find \(a\):
  $$
  1 = a(3 - 2)(3 - 4) \implies a = -1
  $$
- Final equation:
  $$
  y = -(x - 2)(x - 4)
  $$

---

#### **3. Center of the ellipse \(x^2 + 4y^2 - 4x - 16y + 16 = 0\):**
- Rearrange and complete the square:
  $$
  (x^2 - 4x) + 4(y^2 - 4y) = -16
  $$
  $$
  (x - 2)^2 + 4(y - 2)^2 = 4
  $$
- Center of the ellipse:
  $$
  (h, k) = (2, 2)
  $$

---

#### **4. Slope (\(m > 0\)) of the tangent line \(y = mx - 5\) to the circle \(x^2 + y^2 = 1\):**
- Circle equation: \(x^2 + y^2 = 1\).
- Substitute \(y = mx - 5\):
  $$
  x^2 + (mx - 5)^2 = 1
  $$
  Expand:
  $$
  x^2 + m^2x^2 - 10mx + 25 = 1
  $$
  Combine terms:
  $$
  (1 + m^2)x^2 - 10mx + 24 = 0
  $$
- Tangency condition (discriminant \(= 0\)):
  $$
  (-10m)^2 - 4(1 + m^2)(24) = 0
  $$
  Simplify:
  $$
  100m^2 - 96 - 96m^2 = 0 \implies 4m^2 = 96 \implies m^2 = 24 \implies m = 2\sqrt{6}/2
  $$
  Tangent slope:
  $$
  m = \sqrt{6}
  $$

---

#### **5. Intersection points of \(x^2 - y^2 = 1\) and \(x^2 + 4y^2 = 6\):**
- From the hyperbola equation \(x^2 = y^2 + 1\), substitute into the ellipse equation:
  $$
  (y^2 + 1) + 4y^2 = 6
  $$
  Combine terms:
  $$
  5y^2 + 1 = 6 \implies 5y^2 = 5 \implies y^2 = 1
  $$
  \(y = \pm 1\). Substitute \(y = 1\) and \(y = -1\) into \(x^2 = y^2 + 1\):
  $$
  x^2 = 1 + 1 = 2 \implies x = \pm \sqrt{2}
  $$
- Intersection points:
  $$
  (\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
  $$

---

#### **6. Distance between branches of \(x^2 - y^2 = 1\):**
- The distance between branches is measured along the conjugate axis.
- Standard equation:
  $$
  x^2/a^2 - y^2/b^2 = 1
  $$
- For \(x^2 - y^2 = 1\), \(a = 1\) and \(b = 1\).
- Distance between branches:
  $$
  2b = 2
  $$
